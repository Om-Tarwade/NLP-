```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger

nltk.download('punkt')
nltk.download('average_peceptron_tagger')

text="The dog is eating banana"
tokens=word_tokenize(text)

pos_tags=nltk.pos_tag(tokens)
print(pos_tags)

#using the regular expression
patterns=[
    (r'.*ing$','VBG'),
    (r'.*ed$','VBD'),
    (r'.*s$','NNS'),
    (r'.*','NN')
]
tagger=RegexpTagger(patterns)
print('regex based tagger:', tagger.tag(tokens))
```

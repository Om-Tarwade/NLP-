```python
#single word sentiment
pos=['good','happy','great']
neg=['bad','sad','worst']

word="good"

if word in pos:
    print("positive")
elif word in neg:
    print("negative")
else:
    print("neural")
```

```python
#sentence sentiment analysis
pos=['good','happy','great']
neg=['bad','sad','worst']

text="this is the good day"
words=text.split()

score=0
for w in words:
    if w in pos: score+=1
    if w in neg: score-=1

if(score>0):
    print("positive")
elif(score<0):
    print("Negative")
else:
    print("neural")
```

```python
#polarity based sentiment analysis
from textblob import TextBlob

text="this is the great day"
polarity=TextBlob(text).sentiment.polarity

if polarity>0:
    print("positive")
elif polarity<0:
    print("Negative")
else:
    print("neutral")
```

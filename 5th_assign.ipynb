```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('average_perceptron_tagger')

text="the quick brown fox is running very fast"
tokens=word_tokenize(text)
pos_tags=nltk.pos_tag(tokens)

grammar= r"""
NP:{<DT>?<JJ>*<NN>}
VP:{<VB.*><RB>?}
ADJP: {<JJ>}
ADVP: {<RB>}
"""

chunk_parser=nltk.RegexpParser(grammar)
tree=chunk_parser.parse(pos_tags)

print(tree)
tree.draw()
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('average_perceptron_tagger')
text="my name is harshada ankush bhure"
tokens=word_tokenize(text)
pos_tags=nltk.pos_tag(tokens)
grammar=r"""
NP:{<DT>?<JJ>*<NN>},
VP:{<VB.*><RB>?}
ADVJ:{<JJ>}
ADVP:{<RB>}
"""
chunk_parser=nltk.RegexpParser(grammar)
tree=chunk_parser.parse(pos_tags)

print(tree)
tree.draw()
```

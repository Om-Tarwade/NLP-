```python
def cyk(grammar, sentence):
    words=sentence.split()
    n=len(words)

    table=[[set() for j in range(n)] for i in range(n)]

    for length in range(2,n+1):
        for i in range(n-length+1):
            j=i+length-1

            for k in range(i,j):
                for A in grammar:
                    for rule in grammar[A]:
                        if isinstance(rule,tuple):
                            B,C=rule
                            if B in table[i][k] and C in table[k+1][j]:
                                table[i][j].add(A)
            return 'S' in table[0][n-1]
        
grammar={
    's':[('NP','VP')],
    'NP':[('Det', 'N')],
    'VP':[('V','NP')],
    'Det':['the'],
    'N':['dog'],
    'V':['sees']
}
print(cyk(grammar, "the dog sees"))
```

```python
text="i love machine learning and i love coding"
words=text.split()

model={}
for i in range(len(words)-1):
    if words[i] not in model:
        model[words[i]]=[]
    model[words[i]].append(words[i+1])

word="love"
if word in model:
    print("next words", model[word])
else:
    print("words not found")
```
